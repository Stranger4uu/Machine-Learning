```python
import pandas as pd
import requests

# Get 100 movie IDs
id_list = []
for i in range(1, 6):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&page={}'.format(i))
    for movie in response.json()['results']:
        id_list.append(movie['id'])

# Loop and collect
df = pd.DataFrame()

for movie_id in id_list:
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key=8265bd1679663a7ea12ac168da84d2e8'.format(movie_id))
    data = response.json()

    temp_df = pd.DataFrame([{
        'id':             data.get('id'),
        'imdb_id':        data.get('imdb_id'),
        'title':          data.get('title'),
        'release_date':   data.get('release_date'),
        'runtime':        data.get('runtime'),
        'budget':         data.get('budget'),
        'vote_average':   data.get('vote_average'),
    }])

    df = pd.concat([df, temp_df], ignore_index=True)

df.shape
```
