```python
# Univariate variable - numerical
```

```python
data = {
    "Salary" : [25000, 30000, 24000, 12000, 10000, 50000, 70000, 64000, 36000, 34500]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
# Line Plot:

plt.plot(df['Salary'], color = "red", marker = ".", linestyle =":", linewidth = 2)
plt.grid()
plt.show()
```

```python
 # Histogram
 plt.hist(df['Salary'], bins = 5)
 plt.grid()
 plt.show()
```

```python
# Box Plot
plt.boxplot(df['Salary'])
plt.grid()
plt.show()
```

```python
df.loc[10] = [0]
```

```python
df.drop(index = 10, inplace = True)
```

```python
# Univariate: Categorical
```

```python
df["dept"] = ['HR', 'IT', 'Finance', 'HR', 'Finance', 'IT', 'IT', 'HR', 'IT', 'HR']
df.head()
```

```python
df.info()
```

```python

# Pie chart
count = df['dept'].value_counts()
count
```

```python
plt.pie(count, labels = count.index, autopct="%1.1f")
plt.show()
```

```python
# count plot
plt.bar(count.index, count, color = ["green", "red", "yellow"])
plt.show()
```

```python
# Bivariate - numerical - numerical
```

```python
df["Age"] = [34, 24, 56, 18, 27, 20, 53, 46, 37, 44]
df.head()
```

```python
# Scatter Plot:
plt.scatter(df['Salary'], df['Age'], color = "black")
plt.grid()
plt.show()
```

```python
sorted_age = df.sort_values("Age")
sorted_age.head()
```

```python
# Line plot
plt.plot(df['dept'], sorted_age['Age'], df['Salary'],  color = "red", marker = ".", linestyle =":", linewidth = 2)
plt.grid()
plt.show()

```

```python
# Bar chart
plt.bar(sorted_age['Age'], df['Salary'], color = ["green", "red", "yellow"])
plt.show()
```

```python
# Bivariate - numerical - categorical
hr_sal = df[df['dept'] == 'HR']['Salary']
hr_sal
```

```python
it_sal = df[df['dept'] == 'IT']['Salary']
it_sal
```

```python
finacne_sal = df[df['dept'] == 'Finance']['Salary']
finacne_sal
```

```python
# Box plot:
plt.boxplot([hr_sal, it_sal, finacne_sal], label =["HR", "IT", "Finance"], labels=["HR", "IT", "Finance"])
plt.legend()
plt.grid()
plt.show()
```

```python
# Pie chart
salary_by_dept = df.groupby('dept')['Salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct="%1.1f", shadow = True, explode = [0.1, 0.1, 0.1])
plt.axis("equal")
plt.show()
```

```python
# Bar plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
finance_mean = sum(finacne_sal)/len(finacne_sal)
```

```python
plt.bar(["HR", "IT", "Finance"], [hr_mean, it_mean, finance_mean], color = ["red", "yellow", "black"])
plt.grid()
plt.show()
```

```python
# Multivariate Analysis: 3 numerical columns

df["Experience"] = [1, 2, 0, 1, 3, 2, 5, 1, 1, 4]
df.head()
```

```python
# Bubble Plot:
plt.scatter(df['Salary']*40, df['Age']*5, s = df['Experience']* 25, color = "pink", edgecolors = "black")
plt.title("Bubble Plot")
plt.xlabel("Salary")
plt.ylabel("Age")

plt.show()
```

```python
# Object Oriented API

fig, axs = plt.subplots(1, 3, figsize = (15, 5))

# Line plot
axs[0].plot(sorted_age["Age"], df["Salary"], color = "black", marker = ".", linewidth = "2")
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# Histogram

axs[1].hist(df["Salary"], bins = 5, color = "pink")
axs[1].grid()
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")

# Box Plot:

axs[2].boxplot(df["Salary"])
axs[2].grid()
axs[2].set_title("Box Plot")
axs[2].set_xlabel("Salary")

plt.show()
```
